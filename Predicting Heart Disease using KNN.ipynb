```python
## separating input as X and ouput/target as y from dataset

X = df.drop(columns='HeartDisease')
y = df['HeartDisease']
```

## Scaling Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  
X_scaled = scaler.fit_transform(X)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=10)
```

## Selecting Value for k

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
err_list=[]

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    knn.fit(X_train, y_train)
    y_pred =knn.predict(X_test)
    err_list.append(1-accuracy_score(y_test, y_pred))
    
err_list
```

```python
## Elbow Method: 

plt.title('Elbow Method')
plt.plot(range(1,21),err_list)
plt.xlabel('K values')
plt.ylabel('Error')
plt.show()
```

```python
## we can see here k with value 14 has minimum error
```

```python
knn = KNeighborsClassifier(n_neighbors=14, metric='euclidean')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

```
